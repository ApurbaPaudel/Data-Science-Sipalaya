### date time split

```python
import pandas as pd
```

```python
data={
    'Start Time':['2025-02-01','2025-02-02','2025-02-03'],
    'End Time':['2025-02-02','2025-02-03','2025-02-04'],
    'Duration(Hours)':[5,8,6]
}
df=pd.DataFrame(data)
df
```

```python
df['Year']=df['Start Time'].str.split('-').str.get(0)
```

```python
df['Month']=df['Start Time'].str.split('-').str.get(1)
```

```python
df['Day']=df['Start Time'].str.split('-').str.get(2)
```

```python
df
```

```python
df.info()
```

```python
df['Year']=df['Year'].astype(int)
```

```python
df['Month']=df['Month'].astype(int)
```

```python
df['Day']=df['Day'].astype(int)
```

### extract username and the domain name from the email

```python
data={
    'Name':['Alice Johnson','Bob Smith','Charlie Brown','Diana Prince'],
    'Email':['alice.johnson@example.com','bob.smith@example.com','cahrlie.brown@example.com','diana.prince@example.com']
}
df=pd.DataFrame(data)
df
```

```python
df['First Name']=df['Name'].str.split().str.get(0)
```

```python
df
```

```python
df['Last Name']=df['Name'].str.split().str.get(1)
```

```python
df
```

```python
df['Username']=df['Email'].str.split('@').str.get(0)
df
```

```python
df['Domain Name']=df['Email'].str.split('@').str.get(1)
```

```python
df
```

```python
df['Domain']=df['Domain Name'].str.split('.').str.get(1)
```

```python
df
```

```python
df['Full Name']=df['First Name'] +' ' + df['Last Name']
```
