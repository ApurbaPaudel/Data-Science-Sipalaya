### scatter plot

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,10))
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',style='sex',hue='time',size='size')
plt.title('total bill vs tips')
plt.show()
```
