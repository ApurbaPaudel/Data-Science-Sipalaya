## title,xlabel,ylabel

```python
plt.figure(figsize=(12,3))

plt.plot(df['index'],df['V Kohli'],marker='<',color='red')
plt.plot(df['index'],df['RG Sharma'],marker='<',color='green')


plt.title("Rohit Sharms Vs BVirat Kiohli Carrer Comparison ")
plt.xlabel("season")
plt.ylabel("runs")


plt.show()
```

## linestyle

```python
plt.figure(figsize=(12,3)) # # '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'

plt.plot(df['index'],df['V Kohli'],marker='<',color='red',linestyle='solid')
plt.plot(df['index'],df['RG Sharma'],marker='<',color='green',linestyle='dashed')


plt.title("Rohit Sharms Vs BVirat Kiohli Carrer Comparison ")
plt.xlabel("season")
plt.ylabel("runs")


plt.show()
```

```python
plt.figure(figsize=(12,3)) # # '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'

plt.plot(df['index'],df['V Kohli'],marker='<',color='red',linewidth=3)
plt.plot(df['index'],df['RG Sharma'],marker='<',color='green',linewidth=5)


plt.title("Rohit Sharms Vs BVirat Kiohli Carrer Comparison ")
plt.xlabel("season")
plt.ylabel("runs")


plt.show()
```

## markersize,marker

```python
plt.figure(figsize=(12,3)) # # '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'

plt.plot(df['index'],df['V Kohli'],marker='p',color='red',linewidth=2,markersize=20)
plt.plot(df['index'],df['RG Sharma'],marker='<',color='green',linewidth=2,markersize=20)


plt.title("Rohit Sharms Vs BVirat Kiohli Carrer Comparison ")
plt.xlabel("season")
plt.ylabel("runs")


plt.show()
```

## plt.text()

```python
price = [48000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]


#plt.text(2016,46000,'(2016,46000)')
for i in range(len(year)):
    plt.text(year[i],price[i],str(year[i])+'->'+str(price[i])) #'2015->4800'

plt.plot(year,price)
```

## axvline,axhline

```python
plt.figure(figsize=(12,7))

plt.plot(df['index'].tolist(),df['V Kohli'].tolist(),marker='.',color='red')
plt.plot(df['index'].tolist(),df['RG Sharma'].tolist(),marker='.',color='green')



plt.title("Rohit Sharms Vs Virat Kiohli Carrer Comparison ")
plt.xlabel("season")
plt.ylabel("runs")

#plt.xlim(2010,2014)
#plt.ylim(200,400)
plt.axvline(2012,color='pink')
plt.axhline(600,color='pink')
plt.xticks(df['index'].tolist())
for i in range(len(season)):
    plt.text(season[i],vk[i],vk[i],bbox=dict(facecolor='red',alpha=1))
    plt.text(season[i],rs[i],rs[i],bbox=dict(facecolor='yellow',alpha=1))




plt.savefig(r'C:\Users\DELL\Desktop\Sipalaya Data Science\datas')


plt.show()
```
