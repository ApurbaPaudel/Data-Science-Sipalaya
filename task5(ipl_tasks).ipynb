## 7.Is there a relationship between umpires and finals?

```python
finals = df[df['venue'].str.contains('Final', case=False, na=False)]

# Count how many finals each umpire officiated
finals_umpire1 = finals['umpire1'].value_counts()
finals_umpire2 = finals['umpire2'].value_counts()
finals_umpire3 = finals['umpire3'].value_counts()

# Combine the umpire counts from all three umpire columns
umpire_counts = finals_umpire1.add(finals_umpire2, fill_value=0).add(finals_umpire3, fill_value=0)

# Display the result: which umpires officiated most finals
umpire_counts = umpire_counts.reset_index()
umpire_counts.columns = ['Umpire', 'Finals Officiated']

# Sort by the number of finals officiated in descending order
umpire_counts_sorted = umpire_counts.sort_values(by='Finals Officiated', ascending=False)

# Display the sorted results
umpire_counts_sorted
```

## 17.How many matches did each team win at home?(based on venue)

```python
df_wins = df[df['result'] == 'normal']

wins_at_home = df_wins.groupby(['winner', 'venue']).size().reset_index(name='wins_at_home')

wins_at_home = wins_at_home[wins_at_home['winner'] == wins_at_home['venue']]

total_wins_at_home = wins_at_home.groupby('winner')['wins_at_home'].sum().reset_index()

total_wins_at_home
```

## 7.Which team has defended handsomely after winning toss i.e won by big margins?

```python
df_field_first = df[df['toss_decision'] == 'field']

big_margin_threshold_runs = 50  
big_margin_threshold_wickets = 5  

df_big_margin_wins = df_field_first[
    (df_field_first['win_by_runs'] > big_margin_threshold_runs) | 
    (df_field_first['win_by_wickets'] > big_margin_threshold_wickets)
]

big_margin_wins_by_team = df_big_margin_wins.groupby('winner').size().reset_index(name='big_margin_wins')

big_margin_wins_by_team_sorted = big_margin_wins_by_team.sort_values(by='big_margin_wins', ascending=False)
```

```python
big_margin_wins_by_team_sorted
```

## 8.Which teams has chased downs totals successfully with good number of wickets left at hand?

```python
df_chasing_successful = df[(df['toss_decision'] == 'bat') & (df['result'] == 'normal')]

wickets_left_threshold = 5  

df_good_wickets_left_chases = df_chasing_successful[df_chasing_successful['win_by_wickets'] >= wickets_left_threshold]

successful_chases_by_team = df_good_wickets_left_chases.groupby('winner').size().reset_index(name='successful_chases')

successful_chases_by_team_sorted = successful_chases_by_team.sort_values(by='successful_chases', ascending=False)

successful_chases_by_team_sorted
```

## 9.Have teams started to score more easily over the years?

```python
import matplotlib.pyplot as plt
df['total_runs'] = df['win_by_runs'] + df['win_by_wickets']  # Assuming this represents the total runs scored in the match

avg_runs_per_season = df.groupby('season')['total_runs'].mean().reset_index(name='avg_runs')

plt.figure(figsize=(10, 6))
plt.plot(avg_runs_per_season['season'], avg_runs_per_season['avg_runs'], marker='o', color='b')
plt.title('Average Runs Scored Per Match Over the Years')
plt.xlabel('Season')
plt.ylabel('Average Runs')
plt.grid(True)
plt.show()
```
