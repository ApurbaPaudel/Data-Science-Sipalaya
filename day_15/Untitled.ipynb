# show the line charts of total runs in each over

```python
sun=df[df['batting_team']=='Sunrisers Hyderabad']
rcb=df[df['batting_team']=='Royal Challengers Bangalore']
```

```python
sun.head()
```

```python
sun.shape
```

```python
rcb.shape
```

```python
rcb.groupby('over')['total_runs'].sum()
```

```python
total_runs=pd.DataFrame(rcb.groupby('over')['total_runs'].sum()).reset_index()  #rcb
total_runs.rename(columns={'total_runs':'rcb'},inplace=True)
total_runs
```

```python
total_runs['sun']=sun.groupby('over')['total_runs'].sum().values
```

```python
total_runs
```

```python
plt.figure(figsize=(12,4))
plt.plot(total_runs['over'],total_runs['rcb'],marker='<',color='green',markersize=8)
plt.xticks(total_runs['over'])
for i in range(len(total_runs['over'].tolist())):
    plt.text(total_runs['over'].tolist()[i],total_runs['rcb'].tolist()[i],total_runs['rcb'].tolist()[i])
plt.xlabel('over')
plt.ylabel('runs')
plt.show()
```

```python
total_runs['over'].tolist()
```

```python
plt.figure(figsize=(12,4))
plt.plot(total_runs['over'],total_runs['sun'],marker='<',color='purple',markersize=8)
plt.xticks(total_runs['over'])
for i in range(len(total_runs['over'].tolist())):
    plt.text(total_runs['over'].tolist()[i],total_runs['sun'].tolist()[i],total_runs['sun'].tolist()[i])
plt.xlabel('over')
plt.ylabel('runs')
plt.show()
```

# Comparing both overs and runs

```python
plt.figure(figsize=(12,4))
plt.plot(total_runs['over'],total_runs['sun'],marker='<',color='purple',markersize=8)
plt.plot(total_runs['over'],total_runs['rcb'],marker='<',color='green',markersize=8)
plt.xticks(total_runs['over'])
for i in range(len(total_runs['over'].tolist())):
    plt.text(total_runs['over'].tolist()[i],total_runs['sun'].tolist()[i],total_runs['sun'].tolist()[i])
    plt.text(total_runs['over'].tolist()[i],total_runs['rcb'].tolist()[i],total_runs['rcb'].tolist()[i])
plt.xlabel('over')
plt.ylabel('runs')
plt.show()
```

# rolling mean

```python
total_runs['rcb_rolling']=total_runs['rcb'].rolling(window=3).mean()
```

```python
total_runs['sun_rolling']=total_runs['sun'].rolling(window=3).mean()
```

```python
total_runs
```

```python
plt.figure(figsize=(12,4))
plt.plot(total_runs['over'],total_runs['rcb_rolling'],marker='<',color='green',markersize=8)
plt.xticks(total_runs['over'])
for i in range(len(total_runs['over'].tolist())):
    plt.text(total_runs['over'].tolist()[i],total_runs['rcb_rolling'].tolist()[i],total_runs['rcb'].tolist()[i])
plt.xlabel('over')
plt.ylabel('runs')
plt.show()
```
