# Problem-1: Draw a line plot of which, the x-axis is the "Year" and the y-axis is sum of "PM2.5" of two countries Iran and China.

```python
df['Year']=df['Year'].astype(int)
```

```python
temp_df=df[df['Country'].isin(['Iran','China'])]
```

```python
temp_df.head(1)
```

```python
pm25_sum=temp_df.groupby('Year')['PM2.5'].sum().reset_index()
```

```python
pm25_sum
```

```python
plt.figure(figsize=(10,5))
plt.plot(pm25_sum['Year'],pm25_sum['PM2.5'],marker='*',color='purple')
plt.xticks(pm25_sum['Year'])
plt.xlabel('Year')
plt.ylabel('Sum of PM2.5')
plt.title('PM2.5 levels over years for Iran and China')
plt.grid()
plt.show()
```

# Problem-2: Draw a histogram of the column "PM10" of which the y-axis represents the probability (see the documentation how to draw the probability).

```python
plt.figure(figsize=(10,5))
plt.hist(df['PM10'].dropna(),bins=30,density=True,alpha=0.7,color='green',edgecolor='black')
plt.xlabel('PM10')
plt.ylabel('Probability')
plt.title('Histogram of PM10 with Probability')
plt.grid()
plt.show()
```

# Problem-3: Draw a scatter plot where x-axis represents "PM2.5" and y-axis represents "PM10" for two countries Poland and Chile.

```python
poland=df[df['Country']=='Poland']
chile=df[df['Country']=='Chile']
```

```python
plt.figure(figsize=(10,5))
plt.scatter(poland['PM2.5'],poland['PM10'],alpha=0.7,color='red',label='poland')
plt.scatter(chile['PM2.5'],chile['PM10'],alpha=0.7,color='green',label='chile')
plt.xlabel('PM2.5')
plt.ylabel('PM10')
plt.title('Scatter plot of PM2.5 and PM10 of Poland and Chile')
plt.legend(loc='upper center')
plt.grid()
plt.show()
```

# Problem-4: Draw a pie chart of top 5 most frequent countries

```python
top=df['Country'].value_counts().sort_values(ascending=False).head(5)
explode=[0.01,0.2,0,0.2,0.3]
```

```python

plt.pie(top,labels=top.index,autopct='%1.1f%%',shadow=True,explode=explode,colors=['pink', 'green', 'blue', 'purple', 'cyan'])
plt.show()
```

# Problem-7 Multiline Plot of all products sales,Give different styes for each products,Add legend at top right.

```python
line_styles={
    'facecream':('solid','orange'),
    'facewash':('dashed','red'),
    'toothpaste':('dotted', 'green'),
    'bathingsoap':('dashdot','purple'),
    'shampoo':((0, (3, 1, 1, 1)),'blue'),
    'moisturizer':((0, (5, 2)),'brown')
}

    
```

```python
plt.figure(figsize=(12,6))
for product,(style,color) in line_styles.items():
    plt.plot(sales['month_number'],sales[product],linestyle=style,color=color,marker='o',label=product)

plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Monthly Sales of All Products')    
plt.legend(loc='upper right')
plt.show()
```

# Problem-8 Show Quarter wise Sales data for all products as multi Bar chart.

```python
sales['quarter'] = (sales['month_number'] - 1) // 3 + 1
```

```python
quarterly_sales = sales.groupby('quarter')[['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']].sum()
```

```python
quarterly_sales.plot(kind='bar', figsize=(10, 6),color=['blue','green','purple','pink','orange','brown'])
plt.xlabel('Quarter')
plt.ylabel('Total Units Sold')
plt.title('Quarterly Sales Data for All Products')
plt.legend(title='Products')
plt.show()
```
