# Extract the name , age , Date of Birth Birth place and Nationality and make new columns respective to their details

```python
df['Name']=df['player_info'].str.split('Name\n').str[1].str.split('\n').str[0]
```

```python
df['DOB']=df['player_info'].str.split('Birth Date\n').str[1].str.split('\n').str[0]
```

```python
df['Birth Place']=df['player_info'].str.split('Birth Place\n').str[1].str.split('\n').str[0]
```

```python
df['Nationality']=df['player_info'].str.split('Nationality\n').str[1]
```

```python
df[['Name','DOB','Birth Place','Nationality']]
```

# Extract the Role , Batting style, Bowling Style , Team and Debut date and make New Columns

```python
df['Role']=df['profesional_info'].str.split('Role\n').str[1].str.split('\n').str[0]
```

```python
df['Batting Style']=df['profesional_info'].str.split('Batting Style\n').str[1].str.split('\n').str[0]
```

```python
df['Bowling Style']=df['profesional_info'].str.split('Bowling Style\n').str[1].str.split('\n').str[0]
```

```python
df['Team']=df['profesional_info'].str.split('Team\n').str[1].str.split('\n').str[0]
```

```python
df['Debut Date']=df['profesional_info'].str.split('Debut Date\n').str[1]
```

```python
df=df.drop('Debut Name',axis=1)
```

```python
df=df.replace('-- : --','Not Available')
```

```python
df=df.dropna().reset_index(drop=True)
```

```python
df['Age']=df['player_info'].str.split('Age\n').str[1].str.split('\n').str[0].str.split(' ').str[0]
```

# Change the date of birth and debut date in date time format and age in numerical columns

```python
df['DOB'] = pd.to_datetime(df['DOB'], errors='coerce')
df['Debut Date'] = pd.to_datetime(df['Debut Date'], errors='coerce')
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
```
