```python
df=student_list.merge(student_response,on='roll_number',how='inner')
```

```python
df = df.merge(correct_answer, on=['question_paper_code', 'question_number'], how='inner')
```

```python
df
```

```python
df = df.merge(question_paper_code, left_on='question_paper_code', right_on='paper_code', how='inner')
```

```python
df
```

```python
df['Math_correct'] = ((df['subject'] == 'Math') & (df['option_marked'] == df['correct_option'])).astype(int)
df['Math_wrong'] = ((df['subject'] == 'Math') & (df['option_marked'] != df['correct_option']) & (df['option_marked'].notna())).astype(int)
df['Math_yet_to_learn'] = ((df['subject'] == 'Math') & (df['option_marked'] == 'E')).astype(int)
```

```python
df['Science_correct'] = ((df['subject'] == 'Science') & (df['option_marked'] == df['correct_option'])).astype(int)
df['Science_wrong'] = ((df['subject'] == 'Science') & (df['option_marked'] != df['correct_option']) & (df['option_marked'].notna())).astype(int)
df['Science_yet_learn'] = ((df['subject'] == 'Science') & (df['option_marked'] == 'E')).astype(int)
```

```python
df['Math_score'] = df['Math_correct']
df['Science_score'] = df['Science_correct']
```

```python
total_math_questions = df[df['subject'] == 'Math']['question_number'].nunique()  
total_science_questions = df[df['subject'] == 'Science']['question_number'].nunique() 
```

```python
df['Math_percent'] = (df['Math_score'] / total_math_questions)*100
df['Science_percent'] = (df['Science_score'] / total_science_questions)*100
```

```python
df['Math_percentage'] = df['Math_percent'].round(0).astype(int)
df['Science_percentage'] = df['Science_percent'].round(0).astype(int)
```

```python
result_df = df.groupby(['roll_number', 'student_name', 'class_y', 'section', 'school_name'], as_index=False).agg({
    'Math_correct': 'sum',
    'Math_wrong': 'sum',
    'Math_yet_to_learn': 'sum',
    'Math_score': 'sum',
    'Math_percentage': 'sum',
    'Science_correct': 'sum',
    'Science_wrong': 'sum',
    'Science_yet_learn': 'sum',
    'Science_score': 'sum',
    'Science_percentage': 'sum'
})
```
