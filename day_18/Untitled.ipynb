```python
from selenium import webdriver  #chrome driver
from selenium.webdriver.common.by import By #element locate
```

```python
import pandas as pd
import numpy as np
import warnings
import time
warnings.filterwarnings('ignore')
```

# load the website

```python
driver=webdriver.Chrome()
driver.get('https://sipalaya.com/all-courses-2/')
```

# extract the data

```python
driver.find_element(By.XPATH,f'/html/body/div[1]/div/div[1]/div/div/div[2]/div/div[12]/div[3]/h3/a').text
```

## loop

```python
course_title=[]
course_duration=[]
course_type=[]
old_price=[]
new_price=[]
for i in range(1,13):
    course=driver.find_element(By.XPATH,f'/html/body/div[1]/div/div[1]/div/div/div[2]/div/div[{i}]/div[3]/h3/a').text
    course_title.append(course)

    course_d=driver.find_element(By.XPATH,f'/html/body/div[1]/div/div[1]/div/div/div[2]/div/div[{i}]/div[3]/div[2]/div/span[2]').text
    course_duration.append(course_d)

    course_t=driver.find_element(By.XPATH,f'/html/body/div[1]/div/div[1]/div/div/div[2]/div/div[{i}]/div[3]/div[3]/div[2]/a[2]').text
    course_type.append(course_t)

    old=driver.find_element(By.XPATH,f'/html/body/div[1]/div/div[1]/div/div/div[2]/div/div[{i}]/div[4]/div/div[1]/span/del/span').text
    old_price.append(old)

    new=driver.find_element(By.XPATH,f'/html/body/div[1]/div/div[1]/div/div/div[2]/div/div[{i}]/div[4]/div/div[1]/span/ins/span').text
    new_price.append(new)
print('success')
    

    
    
```

# make a dataframe

```python
dicts={'course_title':course_title,'course_duration':course_duration,
       'course_type':course_type,'old_price':old_price,'new_price':new_price}
dicts
```

# image url

```python
image_url=[]
course_link=[]
for i in range(1,13):
    img_element=driver.find_element(By.XPATH,f'/html/body/div[1]/div/div[1]/div/div/div[2]/div/div[{i}]/div[1]/a/div/img')
    src_value=img_element.get_attribute('src')
    image_url.append(src_value)

    course_element=driver.find_element(By.XPATH,f'/html/body/div[1]/div/div[1]/div/div/div[2]/div/div[{i}]/div[3]/h3/a')
    course_value=course_element.get_attribute('href')
    course_link.append(course_value)
    

```

```python
dicts={'course_title':course_title,'course_duration':course_duration,
       'course_type':course_type,'old_price':old_price,'new_price':new_price,'image_url':image_url,'course_link':course_link}
dicts
```

```python
df=pd.DataFrame(dicts)
```

```python
df
```

```python
df.to_csv('sipalaya_course.csv',index=False)
```
